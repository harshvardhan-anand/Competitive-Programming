```python
 adj = {
     'a':['b', 'd'],
     'b':['a','c'],
     'c':['b'],
     'd':['a','e', 'f'],
     'e':['d', 'f', 'g'],
     'f':['d', 'e', 'h'],
     'h':['f', 'g'],
     'g':['h']
 }
```

```python
from queue import Queue
```

```python
visited = dict()
parent = dict()
level = dict()
bfs = []  # BFS Traversal
q = Queue()
```

```python
s = 'a'
level[s] = 0
parent[s] = None
q.put(s)

while not q.empty():
    u = q.get()
    visited[u] = True
    bfs.append(u)
    
    for v in adj.get(u):
        try:
            visited[v] # if v is visited then do nothing
        except:
            visited[v] = True
            parent[v] = u
            level[v] = level[u]+1
            q.put(v)

bfs
```

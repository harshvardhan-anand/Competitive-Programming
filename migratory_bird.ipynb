```python
n=int(input())
a=input().split()

ue=[]
counter=[]
do_not_append=0
count=0
switch=1
ue_printer=[]
stop=0

ue.append(a[0])
for i in a:
    i=int(i)
    for k in ue:
        k=int(k)
        if i==k:
            do_not_append=1
    if do_not_append==0:
        ue.append(i)
    do_not_append=0
    
for k in ue:
    k=int(k)
    for i in a:
        i=int(i)
        if k==i:
            count+=1
    counter.append(count)
    count=0
i=0
maxi=counter[0]
while i<len(counter):
        if switch:
            if counter[i]>maxi:
                maxi=counter[i]
        if switch==0:
            if maxi==counter[i]:
                ue_printer.append(ue[i])
        if i==(len(counter)-1) and stop==0:
            switch=0
            i=0
            stop=1
        i+=1
print(min(ue_printer))
```
